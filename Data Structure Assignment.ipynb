```python
# Explain the key features of lists in python

Key Features of lists are

1) Ordered:- Lists maintain the order of elements as they are inserted. The first element added will be at index 0, the second at index 1, and so on.
2) Mutable:- Lists can be modified after creation. We can add, remove, or change elements within a list.
3) Dynamic:- Lists can grow or shrink in size as needed. we can add or remove elements without having to predefine the lists capacity
4) Iterable:- You can iterate over the elements of a list using loops like for or list comprehensions.
5) Nestable:- Lists can contain other lists as elements, creating nested structures. This allows for representing complex data hierarchies.
6) Slicing:- We can extract a portion of a list using slicing to create a new list.
```

```python
# Compare and contrast Tuples and List with example

Tuples

1) Tuples are immutable meaning no changes can be made 
2) The implication of iterations is comparatively Faster
3) A Tuple data type is appropriate for accessing the elements
4) Tuple consumes less memory as compared to the list
5) Tuple does not have many builtin methods

List

1) Lists are mutable
2) The implication of iterations is Time consuming
3) The list is better for performing operations, such as insertion and deletion.
4) Lists consume more memory
5) Lists have several builtin methods

# Example for list

List = [1, 2, 4, 4, 3, 3, 3, 6, 5]
print("Original list ", List)

List[3] = 77
print("Example to show mutability ", List)

# Example for Tuple

tuple1 = (0, 1, 2, 3)
tuple1[0] = 4
print(tuple1)
```

```python
# Describe how to add modify and delete items in a dictionary with examples

A Python dictionary is a data structure that stores the value in key:value pairs.

Dict = {}
print("Empty Dictionary: ")
print(Dict)
Dict[0] = 'Ashish'
Dict[2] = 'jarial'
Dict[3] = 1
print("\nDictionary after adding 3 elements: ")
print(Dict)

Dict['Value_set'] = 2, 3, 4
print("\nDictionary after adding 3 elements: ")
print(Dict)

Dict[2] = 'Welcome'
print("\nUpdated key value: ")
print(Dict)
Dict[5] = {'Nested': {'1': 'Saurabh', '2': 'Shubham'}}
print("\nAdding a Nested Key: ")
print(Dict)
```

```python
# Discuss the importance of dictionary keys being immutable and provide examples

In dictionaries, immutable keys are crucial for efficient value lookup and
maintaining consistent key-value relationships. If keys were mutable, their
hash values could change, making it difficult or impossible to find the associated values

Dictionaries use a hash table implementation that relies on the keys hash value for fast
retrieval. If the key could change, its hash value might also change, causing collisions
and slowing down lookups

Immutability guarantees that a key always points to the same value, preventing
unexpected changes and ensuring data integrity

Since immutable objects cant change their value, Python can optimize memory usage
by storing them more efficiently.

Immutability provides inherent thread safety, as concurrent access to the key wont result in unexpected modifications.
```
